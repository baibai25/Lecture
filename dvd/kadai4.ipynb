```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.naive_bayes import GaussianNB
sns.set_style('whitegrid')
```

```python
df1 = pd.DataFrame([[25, 'sports', 'L'], [20, 'vintage', 'H'], [25, 'sports', 'L'],
                   [45, 'suv', 'H'], [20, 'sports', 'H'], [25, 'suv', 'H']], columns=['Age', 'Car', 'Class'])
df1
```

```python
name = ['Car', 'Class']

for i in name:
    le = LabelEncoder()
    le = le.fit(df1[i])
    df1[i] = le.transform(df1[i])
df1
```

```python
X1 = np.array(df1[['Age', 'Car']])
y1 = np.array(df1['Class'])
```

```python
clf1 = GaussianNB()
clf1.fit(X1, y1)
clf1.predict([[23, 3]]) #[23, truck]
```

```python
df2 = pd.DataFrame([['T', 'T', 5.0, 'Y'], ['T', 'T', 7.0, 'Y'], ['T', 'F', 8.0, 'N'], ['F', 'F', 3.0, 'Y'], ['F', 'T', 7.0, 'N'], 
                   ['F', 'T', 4.0, 'N'], ['F', 'F', 5.0, 'N'], ['T', 'F', 6.0, 'Y'], ['F', 'T', 1.0, 'N']],
                  columns=['a1', 'a2', 'a3', 'Class'])
df2
```

```python
name = ['a1', 'a2', 'Class']

for i in name:
    le = LabelEncoder()
    le = le.fit(df2[i])
    df2[i] = le.transform(df2[i])
df2
```

```python
X2 = np.array(df2[['a1', 'a2', 'a3']])
y2 = np.array(df2['Class'])
```

```python
clf2 = GaussianNB()
clf2.fit(X2, y2)
clf2.predict([[1, 0, 1.0]]) #[T, F, 1.0]
```

+ Q1: (Age=23, Car=truck) is classified as Class H.
+ Q2: (T, F, 1.0) is classified as Class N.
